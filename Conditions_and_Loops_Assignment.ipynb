Q1. What are conditional statements? Explain conditional statements with syntax and examples.

- Conditional statements are used to do a certain task based on conditions. For example, if number is even do this or if a number is odd do that.
Syntax:
```
if(condition){
  // Do something if condition is true
}else{
  // Do something if condition is false
}
```
Example:
```
let x = 10; // input
if(x % 2 === 0){ // Checks if number is fully divisible by 2
  console.log("Even"); // if yes
}else{
  console.log("Odd"); // if no
}
```

2. Write a program that grades students based on their marks.
  *   If greater than 90 then A Grade
  *   If between 70 and 90 then a B grade
  *   If between 50 and 70 then a C grade
  *   Below 50 then an F grade

- Below is the given program for this.

```
let marks = 55; // Update your marks for different test cases
let grade;

if (marks >= 90) {
  grade = "A";
} else if (marks >= 70) {
  grade = "B";
} else if (marks >= 50) {
  grade = "C";
} else {
  grade = "F";
}

console.log(`You got Grade ${grade} for ${marks} Marks.`);
```

3. What are loops, and what do we need them? Explain different types of loops with their syntax and examples.
- Loops are somethings that repeats, We need loops when we need to repeat some task for a definite or indifinite amount of time or even until a statement is true.
There are three types of loops in javascript.
  *   i. While Loop: It runs until a condition is true. For example, following program will print my name 10 times:
Syntax:
```
while(condition){
  // Do something
}
```
Example:
```
let x = 0;
while(x < 10){
  console.log("Manish Kumar Verma");
  x++;
}
```

  *   ii. Do-While Loop: Similar to while loop, but it will execute at least once regradless of condition. Syntax:
```
do{
  // task
}while(condition)
```
Example: it will print my name, even if the condition is false
```
do{
  console.log("Manish Kumar Verma");
}while(1 < 0)
```
  *   iii. For loop: Its loops when the numbers of iterations are known to us. It includes a initialization of variable followed by condition check and then update the value (increment/decrement).
  Syntax:
```
for(initialization; condition; increment){
  // do something
}
```
 Example: it will print my name 100 times;
```
for(let i = 0; i < 100; i++){
  console.log("Manish Kumar Verma");
}
```

Q5. Use the while loop to print numbers from 1 to 25 in ascending and descending order.
- Ascending Order:
```
let i = 1;
while(i <= 25){
  console.log(i);
  i++;
}
```
Decending Order:
```
let i = 25;
while(i > 0){
  console.log(i);
  i--;
}
```
